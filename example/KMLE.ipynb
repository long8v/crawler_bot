```python
for text, url in sub_2:
    print(url)
    sub_3[text] = sub_pages(url, visited=set([]), start_root=False)[0]
```

```python
term_dict = {}
```

```python
for alphabet, term_tuple in sub_3.items():
    for term, url in term_tuple:
        term_dict[term] = parsing(url).find('p').text.replace(u"\xa0", " ")       
```

```python
term_df['term'] = [str(k).split(':')[0] for k in term_dict.keys()]
term_df['aka'] = [str(k).split(':')[-1] for k in term_dict.keys()]
term_df['discription'] = term_dict.values()
```
